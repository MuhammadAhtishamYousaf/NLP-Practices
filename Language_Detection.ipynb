### Preprocessed the text data using CountVectorizer.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer()

text_vectors=vectorizer.fit_transform(x)

#len of unique words
len(vectorizer.get_feature_names_out())
```

### Split the data into training and testing sets.

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(text_vectors,y,test_size=0.20,random_state=42)


print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Trained a Multinomial Naive Bayes model on the training data.

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()

model.fit(x_train,y_train)
```

### Evaluated the model's accuracy on the testing data.

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_true=y_test,y_pred=y_pred)
accuracy
```
